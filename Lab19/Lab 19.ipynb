<h2>Introduction</h2>
Last lab we looked at the ground state configuration of groups of small charged balls. What we did not do was to determine the forces between the balls. We know that the balls are charged but we do not know if they act as point charges or dipoles or even quadrupoles. We will assume that the electrostatic force between any two balls is<br>
$F = \frac{c}{r^a}$<br>
where c is a constant related to the charges of the balls and a is an integer.  We will try to determine a. If the balls act as point charges, then the force law is Coulomb's law and $a=2$. On the other hand, if the balls are effectively dipoles, then the force goes as $1/r^4$ and so $a=4$. To see that the force between two dipoles goes like $1/r^4$, recall that there is no net force on a dipole in a constant electric field. What counts is the gradient or change in the electric field. The force on a dipole in a electric field $E$ is proportional to the dipole moment $\vec p=q \vec d$, where $q$ is the charge on each end of the dipole and $\vec d$ is the distance separating the positive and negative charges. $\vec p$ is a vector quantity, conventionally taken to point along the axis of the two charges from negative to positive.<br><br>For those of you who have taken vector calculus, we can make this statement precise.  Recall that the electrostatic force on a point charge $q$ is<br>
$\vec F = q \vec E$<br>
For a dipole, it is a little more complicated than this.  The net force on a dipole is zero unless the external electric field $\vec E$ is nonconstant. A dipole with dipole moment $\vec p_1$ obeys<br>
$\vec F = (\vec  p_1 \cdot \nabla) \vec E$<br>
The magnitude of the electric field due to a dipole is $ \frac{p_2}{4 \pi \epsilon_0 r^3}$ ($p_2$ may be different from $p_1$ in general), so the force between two dipoles is proportional to $ \frac{p_1 \ p_2}{r^4}$.  If we instead had quadrupoles, then the force would be a $1/r^5$ force, and so forth.<br><br>
In this experiment we will attempt to determine the force law for the force between our charged balls. We will do this by tilting our apparatus slightly and measuring the resulting equilibrium positions of the balls. We will then compare our measured positions with the calculated  positions for an $1/r^a$ force law and thus determine $a$.

<h2>Calculating the Expected Ball Positions</h2>
<h3>Balancing Forces</h3>
Suppose that we tilt the apparatus slightly. What is the force on each ball? Each ball experiences a gravitational force of $m g \sin(\theta)$ where $\theta$ is the (small) angle at which the plate is tilted. Each ball also experiences electrostatic forces from all the other balls. The force from each ball is $c/r^a$ where $c$ is an unknown constant which depends on the charge on each ball. We will assume that all the balls are equally charged. The balls and their numbering are shown below.

```python
#image
```

The ball numbered zero is the ball at the end of the slit. It is essentially fixed in place by the end of the slit. Let $x_i$ be the distance from the ith ball to ball 0. Then the force on ball 1 is<br>
$F_1 = - mg \sin(\theta) + c/x_1^a - c /(x_2 - x_1)^a  - c /(x_3 - x_1)^a - c/(x_4 - x_1)^a - c/(x_5 - x_1)^a - c/(x_6 - x_1)^a$<br>
The force on ball 2 is:<br>
$F_2 = - mg \sin(\theta) + c/x_2^a - c /(x_2 - x_1)^a  - c /(x_3 - x_2)^a - c/(x_4 - x_2)^a - c/(x_5 - x_2)^a - c/(x_6 - x_2)^a$<br>
and so on for balls 3-6.<br><br> If the system is in equilibrium then all these force must be zero. This gives us six equations in six unknowns to solve for the ball positions. These are not easy equatons to solve. We can not solve them analytically. We can solve them numerically.

<h2>The Nearest Neighbor Approximation</h2>
We will start by make what is known as the <u>nearest neighbor approximation</u>. We will assume that each ball is effected only by its nearest neighbors. This should be a good approximation (we will have to check it) if a=4 and a poor approximation if a=1 (can you think of why this is true?)! <br><br>
Suppose we have just four balls (ball 0, 1, 2, and 3). Then in the nearest neighbor approximation our equations are <br>
$-\frac{mg \sin(\theta)}{c} + \frac{1}{x_1^a} - \frac{1}{(x_2 - x_1)^a} = 0$<br>
$-\frac{mg \sin(\theta)}{c} + \frac{1}{(x_2 - x_1)^a} - \frac{1}{(x_3 - x_2)^a} = 0$<br>
$-\frac{mg \sin(\theta)}{c} + \frac{1}{(x_3 - x_2)^a} = 0$<br><br>
We can simplify these equations by a change of variables: Let's rescale all of our distances $x_i \rightarrow x_i(\frac{c}{mg \sin(\theta)})$.  This lets us work with the (now dimensionless!) simpler set of variables, and only worry about these constants at the end.<br><br>
In these units, the equations become.
<br><br>
$-1 + \frac{1}{x_1^a} - \frac{1}{(x_2-x_1)^a} = 0$<br>
$-1 + \frac{1}{(x_2 - x_1)^a} - \frac{1}{(x_3-x_2)^a} = 0$<br>
$-1 + \frac{1}{(x_3-x_2)^a} = 0$<br>
It is easy to solve these equations by inspection:
From the third equation, we see that $x_3 - x_2 = 1$. Then the second equation implies that $x_2 - x_1 = (1/2)^{1/a}$, and then equation one gives $-1 + 1 / x_1 ^a - 2 = 0$  so $x_1 = 3^{-1/a}$.  To summarize:<br><br>
$x_1 = 3^{-1/a}$<br>
$x_2 = 2^{-1/a} + 3^{-1/a}$<br>
$x_3 = 1 + 2^{-1/a} + 3^{-1/a}$<br>
<br>
See the patterns? Continuing along these lines, one can show that for $N_{total}$ balls:
<br>
$x_n = \displaystyle\sum_{i=N_{total}-n}^{N_{total}-1} i^{-1/a}$<br>
where $N_{total}$ is the total number of balls (in the above example, $N_{total} = 4$) and $x_n$ is the $n$th ball position. $x_0 = 0$ by definition.

```python
def x(n_total, n, a):
    return sum(i**(-1/a) for i in range(n_total-n, n_total))
```

```python
n_total = 6
data1 = [x(n_total, n, 1) for n in range(n_total)]
data1
```

if $a=2$ then the positions of balls one through six are

```python
data2 = [x(n_total, n, 2) for n in range(n_total)]
data2
```

if $a=4$ then the positions of balls one through six are

```python
data4 = [x(n_total, n, 4) for n in range(n_total)]
data4
```

```python
import numpy as np
from matplotlib import pyplot as plt

x_axis = np.linspace(0, n_total, 6)
plt.plot(x_axis, data1, label='1/r Force')
plt.plot(x_axis, data2, label='1/r^2 Force')
plt.plot(x_axis, data4, label='1/r^4 Force')
plt.legend()
plt.show()
```

You see that we get quite different curves depending on whether the force law is $1/r$, $1/r^2$ or $1/r^4$. Now we want to normalize the positions so the last ball is at $x=1$

```python
normed_data1 = np.asarray(data1) / data1[-1:]
normed_data2 = np.asarray(data2) / data2[-1:]
normed_data4 = np.asarray(data4) / data4[-1:]
```

We can now plot the normalized ball positions.

```python
plt.plot(x_axis, normed_data1, label='1/r Force')
plt.plot(x_axis, normed_data2, label='1/r^2 Force')
plt.plot(x_axis, normed_data4, label='1/r^4 Force')
plt.legend()
plt.show()
```

<h2>Checking the NN Approximation</h2>
We need to check the nearest neighbor approximation. We expect this approximation to be good for a $1/r^4$ force and poor for $1/r$ but we need to check this.  We wrote down the force equations earlier and they were

```python
from scipy.optimize import fsolve
import numpy as np

def equations(variables, a):
    x1, x2, x3, x4, x5, x6 = variables
    F1 = -1 + 1/x1**a - 1/(x2 - x1)**a - 1/(x3 - x1)**a - 1/(x4 - x1)**a - 1/(x5 - x1)**a - 1/(x6 - x1)**a
    F2 = -1 + 1/x2**a + 1/(x2 - x1)**a - 1/(x3 - x2)**a - 1/(x4 - x2)**a - 1/(x5 - x2)**a - 1/(x6 - x2)**a
    F3 = -1 + 1/x3**a + 1/(x3 - x1)**a + 1/(x3 - x2)**a - 1/(x4 - x3)**a - 1/(x5 - x3)**a - 1/(x6 - x3)**a
    F4 = -1 + 1/x4**a + 1/(x4 - x1)**a + 1/(x4 - x2)**a + 1/(x4 - x3)**a - 1/(x5 - x4)**a - 1/(x6 - x4)**a
    F5 = -1 + 1/x5**a + 1/(x5 - x1)**a + 1/(x5 - x2)**a + 1/(x5 - x3)**a + 1/(x5 - x4)**a - 1/(x6 - x5)**a
    F6 = -1 + 1/x6**a + 1/(x6 - x1)**a + 1/(x6 - x2)**a + 1/(x6 - x3)**a + 1/(x6 - x4)**a + 1/(x6 - x5)**a
    return [F1, F2, F3, F4, F5, F6]

initial_guess = data1

a = 1

solutions1 = fsolve(equations, initial_guess, args=(a))

print(solutions1)
```

```python
initial_guess = data2

a = 2

solutions2 = fsolve(equations, initial_guess, args=(a))

print(solutions2)
```

```python
initial_guess = data4

a = 4

solutions4 = fsolve(equations, initial_guess, args=(a))

print(solutions4)
```

The only remaining step is to normalize the ball positions with the same normalization that we used in the nearest neighbor approximation.

```python
norm_sol1 = np.asarray(solutions1)/data1[-1:]
norm_sol2 = np.asarray(solutions2)/data2[-1:]
norm_sol4 = np.asarray(solutions4)/data4[-1:]
```

We can now plot our exact ball positions:

```python
x_axis = np.linspace(0, n_total, 6)
plt.plot(x_axis, norm_sol1, label='1/r Force')
plt.plot(x_axis, norm_sol2, label='1/r^2 Force')
plt.plot(x_axis, norm_sol4, label='1/r^4 Force')
plt.legend()
plt.show()
```

It will be illustrative to see the difference between the NN approximation with the full solution for each value of a:

```python
plt.plot(x_axis, norm_sol1, label='Full')
plt.plot(x_axis, normed_data1, label='NN')
plt.legend()
plt.show()
```

```python
plt.plot(x_axis, norm_sol2, label='Full')
plt.plot(x_axis, normed_data2, label='NN')
plt.legend()
plt.show()
```

```python
plt.plot(x_axis, norm_sol4, label='Full')
plt.plot(x_axis, normed_data4, label='NN')
plt.legend()
plt.show()
```

As you can see the approximation is extremely good, even for $1/r$
