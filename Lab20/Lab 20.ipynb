The torque on a magnetic dipole is $\vec \tau = \vec \mu \times \vec B$.  On the ball, the magnetic moment $\mu$ is oriented along the handle as shown in Figure 2. By adjusting the current, and thus $B$, the amount of magnetic torque can be adjusted.

Figure 3: There is a small wood or aluminum rod which runs through the handle on the ball. A small plastic weight can be moved along the rod. The gravitation torque is $\vec \tau = \vec r \times m \vec g$ where $m$ is the mass of the plastic weight, and $\vec r$ is the distance from the weight to the center of the ball. The torque (use the right-hand rule again) is into the page (or screen). There is a second gravitation torque due to the fact that the rod is massive. If the rod has length $L$ and mass $M_{rod}$ then the torque due to rod is $\tau_{rod} = (L / 2) M_{rod} g \sin(\theta)$. The factor of 1/2 comes from the fact the we want the center of mass of the rod.

The gravitational torque will cause the ball's handle to rotate downward (since $\tau = dL/dt$ ). The magnetic torque, on the other hand, will cause the ball's handle to rotate upward. If the ball is in equilibrium then these torques must be equal and opposite:
$\tau_{mag} = \tau_g + \tau_{rod}$ or $\mu B \sin(\theta) = r m g \sin(\theta) + (L/2) M_{rod} g \sin(\theta)$ . 

Thus $r = \frac{\mu}{mg} B - \frac{LM_{rod}}{2m}$

Thus $B(r) = \frac{mg}{\mu} r + \frac{L M_{rod} g}{2 \mu}$. Therefore, if  you measure r for a number of B values and plot the result, the slope will be $\frac{mg}{\mu}$ and the intercept will be $\frac{L M_{rod} g}{2 \mu}$. Note that the value of $r$ used in the equation is the distance from the <u>center of the ball</u>. You must therefore add the radius of the ball and the height of the handle to the distance between the handle and the movable weight.

<h2>Taking/Analyzing Data</h2>

Orient the ball and the rod as shown in Figure 1. Turn the air on to minimize friction, making sure that the field orientation is up and the field gradient is off. Start with the weight close to the handle, turn up the current until the system is in equilibrium and record the value of r and the current i. Then move the weight out and repeat the process. Do this for a number of different r values. The relationship between the magnetic field at the center of the ball and the current in the coils is given by the manufacturer as $\frac{B}{i} = 1.36 \pm 0.03 \times 10^-3$ Tesla/Ampere. Some sample data is shown below. The left column is $i$ and right column is $r$ in meters. You will want more data points than are shown here.

```python
data = [
   [0.044, 1.42],
   [0.0509, 1.62],
   [0.0592, 1.8],
   [0.0707, 2.05],
   [0.0779, 2.2],
   [0.0988, 2.8],
   [0.11964, 3.22]
]
```

We can convert the current data into magnetic field data.

```python
 MagneticData = [[i[0], i[1]* (1.36 * 10**-3)] for i in data]
MagneticData
```

Plotting the data gives

```python
from matplotlib import pyplot as plt

x, y = zip(*MagneticData)

plt.plot(x, y, "o")
plt.show()
```

We can use polyfit to fit the data; a linear model seems to work here, as expected.

```python
fit = np.polyfit(x,y, deg = 1)
fit
x_axis = np.linspace(0.044,.12,10)
fit_val = np.polyval(fit, x_axis)

plt.plot(x, y, "o")
plt.plot(x_axis, fit_val)
plt.show()
```

```python
slope = fit[0]
slope
```

From the fit we see that the slope is .0326 T/m and we know that slope =$ \frac{mg}{\mu}$. The mass in this example weighs 0.0014 kg and we find that $\mu$ = 0.42 A-m^2.

```python
m = 0.0014;
g = 9.80;
μ = (m*g)/(slope)
μ
```

The standard deviation of the fit is given by

```python

y_proper_dim = np.polyval(fit, x)

std_deviation = np.sqrt(np.mean((y - y_proper_dim)**2))
std_deviation
```

We can make a plot showing the error bar on our data.

```python
fig, ax = plt.subplots()

ax.errorbar(x, y, yerr=std_deviation, fmt='o',capsize=2)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.errorbar(x, y, yerr=std_deviation, fmt='o',capsize=2)
plt.plot(x_axis, fit_val)
plt.show()
```
